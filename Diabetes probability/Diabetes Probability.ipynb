```python
df = pd.read_csv('diabetes.csv')

# Function to map letters to variable pairs
def get_variable_pair(letter):
    if letter == 'A':
        return 'Glucose and Blood Pressure', 'Glucose', 'BloodPressure'
    elif letter == 'B':
        return 'Glucose and Insulin', 'Glucose', 'Insulin'
    elif letter == 'C':
        return 'Glucose and Age', 'Glucose', 'Age'
    elif letter == 'D':
        return 'Insulin and BMI', 'Insulin', 'BMI'
    elif letter == 'E':
        return 'Insulin and Age', 'Insulin', 'Age'
    else:
        return None, None, None

# Function to calculate the probability of having diabetes
def calculate_probability(user_value1, user_value2, filtered_df, var1, var2):
    # Simple probability calculation (just for demonstration)
    # You should use an appropriate probability model based on your dataset and domain knowledge
    # This is a simple example assuming a threshold for high probability
    threshold_value1 = filtered_df[var1].mean() + filtered_df[var1].std()
    threshold_value2 = filtered_df[var2].mean() + filtered_df[var2].std()

    probability = 0.0
    if user_value1 > threshold_value1 and user_value2 > threshold_value2:
        probability = 0.8  # High probability
    elif user_value1 > threshold_value1 or user_value2 > threshold_value2:
        probability = 0.4  # Medium probability
    else:
        probability = 0.1  # Low probability

    return probability

# Display variable options and prompt the user for input
print("Select a variable pair:")
print("A. Glucose and Blood Pressure")
print("B. Glucose and Insulin")
print("C. Glucose and Age")
print("D. Insulin and BMI")
print("E. Insulin and Age")

# Prompt the user for input
user_choice = input("Enter the letter corresponding to your choice: ").upper()

variable_pair, var1, var2 = get_variable_pair(user_choice)
if variable_pair:
    # Filter the dataset based on the chosen variable pair
    filtered_df = df[[var1, var2]]

    # Prompt the user for input values
    user_value1 = float(input(f"Enter value for {var1}: "))
    user_value2 = float(input(f"Enter value for {var2}: "))

    # Plot the data
    print("Plotting data...")
    plt.scatter(filtered_df[var1], filtered_df[var2], c='blue', alpha=0.6)
    plt.scatter(user_value1, user_value2, c='red', marker='o', label='User Input')
    plt.xlabel(var1)
    plt.ylabel(var2)
    plt.legend()
    plt.title('Data Visualization')
    plt.show()

    # Calculate the probability of having diabetes
    probability = calculate_probability(user_value1, user_value2, filtered_df, var1, var2)
    print(f"Probability of having diabetes: {probability:.2%}")
else:
    print("Invalid choice. Please select a valid letter (A, B, C, D, E).")
```
